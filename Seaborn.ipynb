## Scatter Plot:

- Use: Shows the relationship between two continuous variables.
- Example uses: Analyzing the correlation between two variables, identifying clusters or patterns.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.scatterplot(data=tips,x= 'total_bill',y = 'tip')
```

## Line Plot:

- Use: Displays the trend of one or more variables over a continuous axis.
- Example uses: Visualizing time series data, tracking changes in data over a continuous range.

```python
flights = sns.load_dataset("flights")
sns.lineplot(data=flights,x='year',y='passengers',size=None)
```

```python
flights
```

```python
flights_wide= flights.pivot("year","month","passengers")
```

## Bar Plot:

- Use: Compares categorical variables or shows the distribution of a categorical variable.
- Example uses: Comparing sales figures for different products, displaying the frequency of different categories

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.barplot(data=tips, x="day", y="total_bill")
plt.show()           
```

# HISTOGRAM

Histogram:

Use: Shows the distribution of a continuous variable using bins.
Example uses: Analyzing the distribution of exam scores, understanding the spread of income levels.

```python
sns.histplot(data=tips, x="total_bill", bins=10)
plt.show()
```

Box Plot:

Use: Displays the distribution of a continuous variable across different categories.
Example uses: Comparing the distribution of salaries across different job titles, identifying outliers in data.

```python
sns.boxplot(data=tips, x="day", y="total_bill")
plt.show()
```
